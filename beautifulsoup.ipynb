```
from bs4 import BeautifulSoup
import requests

URL = "https://smgplaza.com/how-to-create-a-python-snake-game-using-pygame/"
req = requests.get(URL)

soupp= BeautifulSoup(req.content,'lxml')

print("\nget value of h2 tag in above URL\n")
heading =soupp.find("h2").text
print(heading)

print("\npull information about image\n")
image=soupp.find("img")
print(image.prettify())


print("\nget specific div tag\n")
headingg = soupp.find('div',class_='blog99-title')
print(headingg)
```

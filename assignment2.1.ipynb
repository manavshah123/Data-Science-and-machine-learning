```
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder



data= {'State': ['Maharestra', 'TN', 'Delhi', 'K','G','U']}
df=pd.DataFrame(data,columns=['State'])

labelencoder = LabelEncoder()

df['state'] = labelencoder.fit_transform(df['State'])
df


color= {'colors': ['red','green','blue','red']}
df1=pd.DataFrame(color,columns=['colors'])
labelencoder = LabelEncoder()

df1['color'] = labelencoder.fit_transform(df1['colors'])
df1
```

```
#************************************************************************#
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

data= {'country': ['Germany', 'India', 'Uk', 'Egypt','Iran'],
       'continent': ['Europe', 'Asia', 'Europe', 'Africa','Asia'],
       'code': ['G', 'I', 'U', 'E','N']}
ex_4=pd.DataFrame(data,columns=['country','continent','code'])

labelencoder = LabelEncoder()
ex_4['Leabeled_continent'] = labelencoder.fit_transform(ex_4['continent'])
ex_4
```

```
import matplotlib.pyplot as plt

x = [0,1,2,3,4]
y = [0,2,4,6,8]

### 1st step
plt.plot(x,y)

plt.title('First Demo Example', fontdict={'fontname':'Arial','fontsize':20})
plt.xlabel('x axis')
plt.ylabel('y axis')

plt.savefig('demo1.png')
```

```
import matplotlib.pyplot as plt

x = [0,1,2,3,4]
y = [0,2,4,6,8]

### labeling
plt.plot(x,y,label='2x')

plt.title('First Demo Example', fontdict={'fontname':'Arial','fontsize':20})
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.savefig('demo1.png')
plt.show()
```

```
import matplotlib.pyplot as plt

x = [0,1,2,3,4]
y = [0,2,4,6,8]

### labeling color and line width
plt.plot(x,y,label='2x', color='red', linewidth=5)

plt.title('First Demo Example', fontdict={'fontname':'Arial','fontsize':20})
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.savefig('demo1.png')
plt.show()
```

```
import matplotlib.pyplot as plt

x = [0,1,2,3,4]
y = [0,2,4,6,8]

### Add marker and marker size and edge color of marker
plt.plot(x,y,label='2x', color='red', linewidth=3 , marker='*' , markersize=15 , markeredgecolor='black')

plt.title('First Demo Example', fontdict={'fontname':'Arial','fontsize':20})
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
plt.savefig('demo1.png')
plt.show()
```

```
import matplotlib.pyplot as plt
from google.colab import drive

x = [0,1,2,3,4]
y = [0,2,4,6,8]

### add value of x and y axis using xticks and yticks 
plt.plot(x,y,label='2x', color='red', linewidth=3 , marker='*' , markersize=15 , markeredgecolor='black')

plt.title('First Demo Example', fontdict={'fontname':'Arial','fontsize':20})
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
images_dir = '/content/drive/MyDrive/Colab Notebooks'
plt.savefig(f"{images_dir}/abc.png")
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y,'b*--')

x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'r',marker='^',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'g--')
plt.title('First Demo Example', fontdict={'fontname':'Arial','fontsize':20})
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
images_dir = '/content/drive/MyDrive/Colab Notebooks'
plt.savefig(f"{images_dir}/abc_dash line.png")
plt.show()
```

```
#change figure size

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(7,5),dpi=200)

x = [0,1,2,3,4]
y = [0,2,4,6,8]

plt.plot(x,y,'b*--')

x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,'r',marker='^',label='x^2')
plt.plot(x2[5:],x2[5:]**2,'g--')
plt.title('First Demo Example', fontdict={'fontname':'Arial','fontsize':20})
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
images_dir = '/content/drive/MyDrive/Colab Notebooks'
plt.savefig(f"{images_dir}/abc_dash line2.png")
plt.show()
```
